```python
from skimage import io
from matplotlib import pyplot as plt
import matplotlib_inline.backend_inline
import matplotlib
matplotlib_inline.backend_inline.set_matplotlib_formats('pdf', 'svg')
matplotlib.style.use('seaborn')
```

```python
from geometrica_recognition import to_1channel
from geometrica_recognition import mask_to_remove_noise
from geometrica_recognition import find_binary_borders
from geometrica_recognition import mask_cards
from geometrica_recognition import clean_figures
from geometrica_recognition import get_figures
from geometrica_recognition import regularized_count
from geometrica_recognition import create_labels
```

```python
def plot(image, labels, contours, fig_count):
    fig, ax = plt.subplots(figsize=(18, 12))
    plt.imshow(image)

    for contour in contours:
        plt.plot(contour[:, 1], contour[:, 0], '-r', linewidth=2)

    props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
    props_main = dict(boxstyle='round', facecolor='wheat', alpha=1)

    for pair, text_label in labels:
        ax.text(pair[1], pair[0], text_label, fontsize=14,
            verticalalignment='top', bbox=props)

    main_label = f'{fig_count} cards'
    ax.text(0, 0, main_label , fontsize=24,
           verticalalignment='top', bbox=props_main)
    ax.axis('off')
```

```python
image = io.imread('samples/IMG_10.jpg')
figures = get_figures(image)
label_lst, contours, fig_count = create_labels(figures)
```

```python
plot(image, label_lst, contours, fig_count)
```
