```python
import numpy as np
from sklearn import datasets
from matplotlib import pyplot as plt
import matplotlib_inline
#matplotlib_inline.backend_inline.set_matplotlib_formats('pdf', 'svg')
```

```python
data = datasets.fetch_openml("mnist_784")
```

```python
d = data.data
y = data.target
```

```python
X_train = d[:60000]
y_train = y[:60000]
X_test = d[60000:]
y_test = y[60000:]
```

```python
features = {}
for f_size in [10, 20, 100]:
    features[f_size] = np.random.randint(0, X_test.shape[1], f_size)
```

```python
import time
```

```python
!pip install tikzplotlib
```

```python
from tikzplotlib import save as tikz_save
```

```python
algs = ['my_own', 'brute', 'kd_tree', 'ball_tree']
yh = np.arange(0)
for i, name in enumerate(algs):
    tm = np.arange(0)
    for size, inds in features.items():
        knn = nn.KNNClassifier(5, name, 'euclidean', False, 5000)
        knn.fit(X_train[:, inds], y_train[inds])
        a = time.time()
        if name == 'my_own':
            knn.find_kneighbors_blockwise(X_test[:, inds], False)
        else:
            knn.__dict__[name].kneighbors(X_test[:, inds], return_distance=False)
        a = (time.time()-a)
        tm = np.append(tm, a)
    if yh.size < 2:
        yh = tm
    else:
        yh = np.vstack((yh, tm))
```

```python
fig = plt.figure(figsize=(15, 6))
ax = fig.add_subplot(111)

plt.title("time efficiency comparison", fontsize=20)
plt.xlabel(r"$k$ value", fontsize=18)
plt.ylabel("Time, s", fontsize=18)

data = np.random.uniform(0, 100, (3,5))
names = [10, 20, 100]

w = 0.2
pos = np.arange(len(names))
ax.set_xticks(pos)
ax.set_xticklabels(names, fontsize=18)
#ax.grid()

ax.bar(pos-w, yh[0], width=w, color="crimson", label="my_own", log=False)
ax.bar(pos,   yh[1], width=w, color="darkcyan",  label="brute", log=False)
ax.bar(pos+w, yh[2], width=w, color="darkorange",  label="kd_tree", log=False)
ax.bar(pos+2*w, yh[3], width=w, color="yellow",  label="ball_tree", log=False)
plt.ylim(-3,230)

ax.legend(loc=1, fontsize=18)

#plt.show()
fig.savefig("timeshist.eps", bbox_inches = 'tight')
#tikz_save("timehist.tex")
```
